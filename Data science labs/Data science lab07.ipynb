```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/titanic_data (4).csv')
```

```python
df.head()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df['Fare']=df['Fare'].replace({'\$':''}, regex=True)
df['Fare']=df['Fare'].astype(float)
df['Fare']=df['Fare'].fillna(df['Fare'].median())
df.drop(columns= ['PassengerId','Ticket'], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder,MinMaxScaler
from sklearn.feature_selection import SelectKBest, chi2
```

```python
categorical_features=['Sex','Embarked']
onehot_encoder=OneHotEncoder(sparse_output=False,drop='first')
onehot_encoded=onehot_encoder.fit_transform(df[categorical_features])
encoded_df=pd.DataFrame(onehot_encoded,columns=onehot_encoder.get_feature_names_out(categorical_features))
data=pd.concat([df.drop(columns=categorical_features),encoded_df],axis=1)
```

```python
data
```

```python
df['FamilySize']=df['SibSp']+df['Parch']+1
df['IsAlone']= (df['FamilySize'] == 1).astype(int)
df.drop(columns=['SibSp','Parch'],inplace=True)
```

```python
df
```

```python
x = df.drop(columns=['Survived','Name','Sex','Embarked'])
y = df['Survived']
```

```python
numerical_features=['Age','Fare','FamilySize']
scaler=MinMaxScaler()
x[numerical_features]=scaler.fit_transform(x[numerical_features])
```

```python
x
```

```python
y
```

```python
selector=SelectKBest(chi2,k=3)
x_selected=selector.fit_transform(x,y)
selected_features=x.columns[selector.get_support(indices=True)]
```

```python
print(selected_features)
```

```python
final_df=pd.DataFrame(x_selected,columns=selected_features)
final_df['Survived']=y
```

```python
final_df
```

```python
x_train,x_test,y_train,y_test=train_test_split(final_df.drop(columns=['Survived']),final_df['Survived'],test_size=0.2,random_state=42)
```
