```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('/content/Mall_Customers.csv')
```

```python
df
```

```python
df.info()
```

```python
x = df.iloc[:, [3, 4]].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss_list= []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss_list.append(kmeans.inertia_)
    print(wcss_list)
plt.plot(range(1, 11), wcss_list)
plt.title('The Elbow Method Graph')
plt.xlabel('Number of clusters(k)')
plt.ylabel('wcss_list')
plt.show()
```

```python
kmeans = KMeans(n_clusters=7, init='k-means++', random_state=42)
y_predict = kmeans.fit_predict(x)
print(y_predict)
```

```python
plt.figure(figsize=(10,5))
plt.scatter(x[y_predict == 0, 0], x[y_predict == 0, 1], s = 50, c = '#AEB727', label = 'Cluster 1')
plt.scatter(x[y_predict == 1, 0], x[y_predict == 1, 1], s = 50, c = '#9527B7', label = 'Cluster 2')
plt.scatter(x[y_predict == 2, 0], x[y_predict == 2, 1], s = 50, c = '#27B795', label = 'Cluster 3')
plt.scatter(x[y_predict == 3, 0], x[y_predict == 3, 1], s = 50, c = '#B72727', label = 'Cluster 4')
plt.scatter(x[y_predict == 4, 0], x[y_predict == 4, 1], s = 50, c = '#27B7B7', label = 'Cluster 5')
plt.scatter(x[y_predict == 5, 0], x[y_predict == 5, 1], s = 50, c = '#B727B7', label = 'Cluster 6')
plt.scatter(x[y_predict == 6, 0], x[y_predict == 6, 1], s = 50, c = '#B7B727', label = 'Cluster 7')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 100, c = 'red', label = 'Centroids')
plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import silhouette_score # Corrected the import statement from sklearn.metrices to sklearn.metrics
silhouette_avg = silhouette_score(x, y_predict)
print(f"Silhouette Average:{silhouette_avg:.2f}")
```
